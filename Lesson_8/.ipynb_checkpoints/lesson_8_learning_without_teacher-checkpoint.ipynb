```python
# обучение без учителя это такие медоты машинного обучения с помощью которых решаются задачи с использованием данных в 
# которых нет правильного ответа. В таких задачах не нужно искать какую то величину а нужно изучить выборку 
# объектов обладающих какими либо определенными признаками и понять как связанны между собой эти объекты или на сколько
# эти объекты похожи.

# по сути нам необходимо разобраться в признаках допустим объектов недвижимости находя например определенны группы таких 
# объектов либо обнаруживая аномалии в данных. Кластеризация один из самых известных алгоритмов.
# С помощью класстеризации ммы можем разбить объекты из имеющихся данных на группы, т.е. на кластеры. Причем если объект
# попадает в какой либо кластер он должен быть похож на другие объекты из этого кластера. Здесь можно связать обучение с 
# учителеи и без него потому что если нам удалось разбить объекты на кластеры, то мы можем использовать информацию о 
# принадлежности объекта к определенному кластеру в качестве обного из признаков определенного объекта и затем исспользовать 
# этот признак в методах обучения с учителем. Либо можно было бы после разделения на кластеры для каждого кластера 
# создать отдельную модель обучения с учителем.


# Класстеризация k-means - k средних
# k-means - один из алгоритмов вероятностго подхода
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3, # какое число кластеров мы хотим получить
               random_state=100)
```

```python
# Если говорить о алгоритме K means то на начальном этапе задается определенное число равное заданное параметру n_clusters
# цельных точек это центроиды - центры классеров. На началной итерации центроиды выбираются случайным образом.
# ближайшие к центроиду точки приписываются определенному кластеру центорм которого эта центроида является
# затем в образовавшихся кластерах образовываются новые центры. туда перемещают центроиды. и все опять повторяется,
# каждой центроиде припиывают ближайшие точки, при этом участие точек меняется кластер если они оказываются ближе к другой 
# центроиде. так происходит до тех пор пока координаты центроид не перестают кардинальным образом меняться
# Так же как и в методе KNN расстояние между объектами в кластерном анализе определеляется как растояние между точками 
# в пространстве признаков. При этом предварительно нужно привести признаки к сходному масштабу и иначе расстояние 
# между точками не будет в равной мере учитывать все признаки.
```

```python
# Функция для создания учебного датасета - признаки рыб обитающих в озере
def make_data(n, seed):
    np.random.seed(seed)
    shift_matrix = np.array([[3, 3], 
                             [6, 9], 
                             [9, 3]])
    
    data = np.random.randn(3, 2, n) + shift_matrix.reshape((3, 2, 1))
    data = np.swapaxes(data, 1, 2)
    data = data.reshape((-1, 2))
    data *= np.array([[20, 0.5]])
    df = pd.DataFrame({'x': data[:, 0], 'y': data[:, 1]},
                     columns=['x', 'y'])
    
    df = df.sample(frac=1.0)
    return df
```

```python
# Понижение размерности данных
# Представим что у нас есть датасет с наблюдениями и мы пытаемся обучить датасет, например дерево решений чтобы научиться
# предсказывать какую то целевую переменную, но к сожалению в данных есть проблема - слишком много признаков у объектов
# Причем выбрать какие либо отдельные признаки не можем так как пострадает точность модели. К тому же может быть такая 
# ситация что у нас слишком много dummy переменных созданных для категориальных признаков , состоящих из множества категорий
# 0 и 1 т.е. мы имеет дело с разряженными данными. Плюс ко всему у нас может быть много признаков которые коррелируют между 
# собой и которые тем не менее мы не можем убрать из модели. Признаков может быть тысячи. на такиом количестве признаков модель
# построить не возможно. Здесь на помощь приходят методы понижения размерности данных. Эти методы как бы сжимают данные,
# если говорить сокращенно они компрессируютс множество признаков в меньшее их количество
# PCA - Principal Component Analysis (Метод главных компонент) 1901 - Карл пирсон
# Для нахождения главных компонент вычисляется ковариационная матрица для имеющихся данных, затем для этой матрицы
# вычисляются собственные векторы и собственные значения. Так же есть метод главных компонент с помощью сингулярного
# разложения матрицы данных. перед применением метода главных компонент признаки должны быть приведены к сходному маштабу
# Это может быть сделано с помощью стандартизации или нормализации данных.
```

```python
# столбец 0 - количество комнат
# столбец 1 - плозать дома (в тысячах квадратных футов)
houses = np.array([[5, 1.6], [4, 1.4], [6, 1.9], [3, 1.1], [4, 1.25]])
```

```python
plt.scatter(houses[:, 0], houses[:, 1])
plt.xlabel('Количество комнат')
plt.ylabel('Площадь дома')
```

```python
# по графику видно что эти 2 признака сильно скоррелированы и их можно расположить вдоль одной прямой, то есть фактически 
# сжать данные до одного измерения
# это означает что данные могут быть представлены одной главной компонентой с минимальной потерей информации
```

```python
# алгоритм PCA
houses
```

```python
# найдем среднее значение каждго признака
mean_values = np.mean(houses, axis=0)
mean_values
```

```python
# отнимем от значений признаков средние значения
houses_centered = houses - mean_values
houses_centered
```

```python
# найдем матрицу ковариаций - это матрица состоязая из попарных ковариаций признаков
covarience_matrix = np.cov(houses_centered.T)
covarience_matrix
```

```python
# проверка правильности найденной матрицы - найдем дисперсии центрированных признаков,  они должны совпадать по значению с 
# главной диагональю ковариационной матрицы
print(np.var(houses_centered[:, 0], ddof=1))
print(np.var(houses_centered[:, 1], ddof=1))
```

```python
# найдем собстевеннве значения eigenvalues и собственные векторы (eigrnvectors) матрицы ковариаций
eigenvalues, eigenvectors = np.linalg.eig(covarience_matrix)
```

```python
eigenvalues
```

```python
eigenvectors
```

```python
# собственные занчения сильно отличаются 1 значение намного больше чем 2 поэтому оставит только собственный вектор из столбца
# 0 массива eigenvectors
# так как мы ищем одну главную компоненту то мы выберем один собственный вектор которому соответствует максимальное 
# собственное значение 1.39445221
# далее мы умножаем центрированные данные на этот вектор и применяем reshape чтобы получится столбец
result_pre = np.dot(houses_centered, eigenvectors[:, 0])
result = result_pre.reshape(-1, 1)
result
```

```python
# таким образом мы получили одну главную компоненту
# Теперь применим класс PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=1)
```

```python
mc = pca.fit_transform(houses)
mc
```

```python
pca.explained_variance_ratio_ # какую долю объясненной дисперсии собержит эта компонента
```

```python
# мы видим что эта доля очень близка к 1 это означает что мы потерям менее 1 процента информации
# Применим PCA в моделях классификации 
```
