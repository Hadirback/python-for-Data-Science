```python
# уберем из модели веса близкие к 0
feats = ['CRIM', 'ZN', 'CHAS', 'NOX', 'RM', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
```

```python
def create_model(X_train, y_train, X_test, y_test, feats, model):
    model.fit(X_train.loc[:, feats], y_train)
    y_pred = model.predict(X_test.loc[:, feats])
    mse = mean_squared_error(y_test, y_pred)
    return mse
```

```python
create_model(X_train_scaled, y_train, X_test_scaled, y_test, feats, LinearRegression())
```

```python
# Модели линейной регрессии с регуляризацией
from sklearn.linear_model import Lasso, Ridge
# величину регуляризации можно менять с помощью alpha - чем выше alpha тем сильнее мы боремся с переобучением.
```

```python
# lasso - линейная регрессия с L1 - регуляризацией
# параметр alpha отвечает за регуляризацию
model = Lasso(alpha=0.003)
```

```python
create_model(X_train_scaled, y_train, X_test_scaled, y_test, feats, model)
```

```python
# Ridge - линейная регрессия с l2 регуляризацией
# параметр alpha отвечает за регуляризацию
model = Ridge(alpha=0.001)
```

```python
create_model(X_train_scaled, y_train, X_test_scaled, y_test, feats, model)
```

```python
# при использовании регуляризации веса признаков понижают свое абсолютное значение. Этим устраняется одна из причин переобучение
# - бесконтрольный рост коэффициентов. Причем чем больше поэффициент alpha тем ближе к 0 располагаются веса. В модели lasso 
# видно что если увеличивать величину альфа коэффициенты некоторых признаков доходят до 0. Такой эффект лассо регрессии
# иногда используют для отбора признаков, так как не значимые признаки обнуляются на меньших значениях альфа
```

```python
n = 21
coeffs = np.zeros((n, len(feats)))
alpha_list = np.logspace(-3, 1, n)
for i, val in enumerate(alpha_list):
    lasso = Lasso(alpha=val)
    lasso.fit(X_train_scaled.loc[:, feats], y_train)
    coeffs[i, :] = lasso.coef_.flatten()
    
for i in range(len(feats)):
    plt.plot(alpha_list, coeffs[:, i])
    
plt.title('убывание абсолютных значений весов признаков при увеличении \n коэффициента регуляризации alpha')
plt.xlabel('alpha')
plt.ylabel('Вес признака')
plt.legend(feats)
```

```python
n = 66
coeffs = np.zeros((n, len(feats)))
alpha_list = np.logspace(-3, 3.5, n)
for i, val in enumerate(alpha_list):
    ridge = Ridge(alpha=val)
    ridge.fit(X_train_scaled.loc[:, feats], y_train)
    coeffs[i, :] = ridge.coef_.flatten()
    
for i in range(len(feats)):
    plt.plot(alpha_list, coeffs[:, i])
    
plt.title('убывание абсолютных значений весов признаков при увеличении \n коэффициента регуляризации alpha')
plt.xlabel('alpha')
plt.ylabel('Вес признака')
plt.legend(feats)
```

```python
# тоность полнота
# Метрика accuracy не пригодна для несбалансированных метрик, т.е. метрик где число объектов одного класса значительно 
# больше чем другого
```

```python
y_valid.value_counts(normalize=True)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_valid, y_pred)
```

```python
#true Negative
TN = ((y_valid == 0) & (y_pred == 0)).sum()
TN
```

```python
#False Positive
FP = ((y_valid == 0) & (y_pred == 1)).sum()
FP
```

```python
#False Negative
FN = ((y_valid == 1) & (y_pred == 0)).sum()
FN
```

```python
#True Postitve
TP = ((y_valid == 1) & (y_pred == 1)).sum()
TP
```

```python
# можно нормализовать матрицу ошибок тогда мы получим значение от доли. и в сумме они должны дать единицу
cm = confusion_matrix(y_valid, y_pred) / y_valid.shape[0]
cm
```

```python
# Точность - доля правильно угаданных выживших от всех тех где мы правлиьно угадали выживших
Precision = TP / (TP + FP)
Precision
```

```python
from sklearn.metrics import precision_score
precision_score(y_valid, y_pred)
```

```python
# следующая метрика полнота - это доля правильно угаданных выживших от тех кто реально выжил
Recall = TP / (TP + FN)
Recall
```

```python
from sklearn.metrics import recall_score
recall_score(y_valid, y_pred)
```

```python
# F1 score - это метрика сочетающая в себе точность и полноту. чем ближе эта метрика к 1 тем лучше модель
F1 = 2 * (Precision * Recall) / (Precision + Recall)
F1
```

```python
from sklearn.metrics import f1_score
f1_score(y_valid, y_pred)
```

```python
# Деревья решений и случайный лес
# Интерпретируемость моделей машинного обучения - как признак влияют на целевую переменную
# Пример интерпретируемости
# в модели линейной регрессии по прогнозированию цен на недвижимость увеличение площади квартиты приводит к увеличению цены
# Применение деревьев решений
# пример использования деревьев решения для задач классификации - определение кредитоспособности заемщика и принятие решения
# о выдаче кредита

# Решения задачи классификации пассажиров титианика с использованием деревьев решений
from sklearn.tree import DecisionTreeClassifier
X_train = pd.read_pickle('X_train.pkl')
y_train = pd.read_pickle('y_train.pkl')

X_valid = pd.read_pickle('X_valid.pkl')
y_valid = pd.read_pickle('y_valid.pkl')
```

```python
max_depth_arr = np.arange(2, 20)
```

```python
max_depth_arr
```

```python
#для деревьев решений нам нет необходимости нормализовывать данные так как когда модель основанная на деревьях разбивает данные
# для разделения на ветви не имеет значение на сколько велика в абсолютном выражении точка разбиения, а важно то насколько
# хорошо выбрана точка чтобы получить максимально отличающиеся друг от друга группы

# максимальная глубина дерева - если мы не обграничим глубину дерева, то модель может слишком точно подстроиться под
# тренировочные данные а на новых данных показать низкую точность т.е. переобучиться. по умолчанию None
accuracy_valid = []
accuracy_train = []
for val in max_depth_arr:
    clf = DecisionTreeClassifier(max_depth=val, random_state=100)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_valid)
    y_pred_train = clf.predict(X_train)
    acc_valid = accuracy_score(y_valid, y_pred)
    acc_train = accuracy_score(y_train, y_pred_train)
    accuracy_valid.append(acc_valid)
    accuracy_train.append(acc_train)
    print('max_depth {} \n\tacc_valid = {} \n\tacc_train = {}\n'.format(val, acc_valid, acc_train))
```

```python
plt.rcParams['figure.figsize'] = 8, 5
plt.plot(max_depth_arr, accuracy_valid)
plt.plot(max_depth_arr, accuracy_train)
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
plt.legend(['valid', 'train'])
plt.xlim(2, 19)
```

```python
#random Forest лучше подойдет и покажет лучшие результаты - в этом алгоритме используется комитет или ансамбль деревьев решений
# для каждого дерева решений создается подвыборка из исходной выборки, причем некоторые экземпляры могут попадать в
# выборку повторно. Такой метод создания подвыборок получил название Бэггинг
# и затем на этих данных строиться дерево решений причем для его обучения используется часть признаков, те которые модель
# посчитала лучшими для этих данных. Получившееся дерево имеет большую глубину и в следствии этого может переобучиться.
# таких деревьем получается несколько и их результаты могут отличаться так что результаты переобучения могут слегка 
# нивилироваться. Например при классификации несколько деревьем случайного леса могут отнести один и тот же объект к разным
# классам. В итоге для отнесения объекта к какому либо классу происходит голосование деревьев и объект относят к тому
# классу за который проголосовало большее кол -во деревьев

from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = [{'n_estimators': [150, 200, 250],
              'max_features': np.arange(5, 9),
              'max_depth': np.arange(5, 10)}]
```

```python
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                  param_grid=parameters,
                  scoring='accuracy',
                  cv=5)
```

```python
# cv - кросс валидация - цифра 5 означает что мы случайным образом разбиваем датасет на 5 частей. По другому эти части 
# называются фолды. и мы стоим модель 5 раз. при первом построении моделей 4 фолда используются в качестве тренировочного
# набора данных т.е. на них модель переобучается. 5 фолд в качестве валидационного. на нам модель оценивается в качестве 
# метрики accuracy/ при втором построении в качестве валидационого фолда используетя другой фолд а старые фолды в качестве 
# тренировочных и так далее. И того для одной модели с одним набором параметров получается 5 значений метрики. и мы 
# усредняем эти значения чтобы у нас было представление о качестве модели
```
