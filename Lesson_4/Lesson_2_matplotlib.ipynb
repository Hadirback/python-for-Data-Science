```python
# Логарифмический масштаб
planet = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
masses = [0.055274, 0.815, 1.0, 0.103, 317.8, 95.0, 14.6, 17.147]

plt.plot(planet, masses)
plt.show()
```

```python
# Применим логарифмический масштаб
masses[planet.index('Earth')] / masses[planet.index('Mercury')]
```

```python
plt.plot(planet, masses)
plt.yscale(value='log')
plt.show()
```

```python
planet_info = pd.Series(masses, index=planet)
planet_info
```

```python
planet_info.plot()
plt.show()
```

```python
# столбчатая диаграмма
planet_info.plot('bar')
plt.show()
```

```python
planet_info.plot(kind='bar', logy=True)
plt.show()
```

```python
planet_info.plot(kind='barh', logx=True)
plt.show()
```

```python
# Визуализация данных на примере индекса Биг-Мака
# Индекс бигмака - неофициальный способ определения паритета покупательной способности

url='http://infographics.economist.com/2018/databank/BMFile2000toJan2018.xls'
# С помощью функции read_excel можно читать данные не только из файлов на компьютере, но и из веб сервера по ссылке

bmi = pd.read_excel(url)
```

```python
# Форматирование текста для графики
x = np.linspace(-5, 5, 101)
x
```

```python
x2 = x**2
x2
```

```python
plt.plot(x, x2)
plt.show()
```

```python
# название графика
plt.plot(x, x2)
plt.title('График f(x)')
plt.show()
```

```python
# Поменяем настройки для большекй четкости графика
# SVG (Scalable Vector Graphics) - масштабируемая векторная графика
%config InlineBackend.figure_format = 'svg'
```

```python
plt.plot(x, x2)
plt.title('График f(x)')
plt.show()
```

```python
plt.plot(x, x2)
plt.title('График f(x)', fontsize=16, fontweight='bold', color = 'DarkRed')
plt.show()
```

```python
plt.plot(x, x2)
plt.title('График f(x)', fontsize=16, fontweight='bold', color = '#808080')
plt.show()
```

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080', 'family': 'Calibri' }
label_dict = {'fontsize': 14, 'color': '#808080', 'family': 'Calibri' }
```

```python
plt.plot(x, x2)
# loc - расположение названия графика - по умолчанию в центре
plt.xlabel('Переменная х', fontdict=label_dict)
plt.ylabel('f(x)', fontdict=label_dict)
plt.title('График f(x)', fontdict=title_dict, loc='left')
plt.show()
```

```python
x3 = x**3
x3
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x3)
plt.plot(x, x2)
# loc - расположение названия графика - по умолчанию в центре
plt.xlabel('Переменная х', fontdict=label_dict)
plt.ylabel('f(x)', fontdict=label_dict)
plt.title('График f(x)', fontdict=title_dict, loc='left')
# добавление легенды
plt.legend(labels=['x2', 'x3'])
plt.show()
```

```python
legend_dict = {'size': 14, 'family': 'Calibri' }
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x3, label='x**3', color='#3bb08f', linestyle='dashed')
plt.plot(x, x2, label='x**2',  linestyle='dotted')
# loc - расположение названия графика - по умолчанию в центре
plt.xlabel('Переменная х', fontdict=label_dict)
plt.ylabel('f(x)', fontdict=label_dict)
plt.title('График f(x)', fontdict=title_dict, loc='left')
# добавление легенды
legend = plt.legend(loc='lower right', prop=legend_dict, frameon=False)
plt.setp(legend.get_texts(), color='grey')
plt.grid(axis='y', color='ghostwhite')
plt.show()
```

```python
# Figure и Axes - основные объекты библиотеки MAtplotlib
# Объекты axes располагаются   внутри figure
x = np.linspace(-5, 5, 101)
x2 = x**2
x3 = x**3
```

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2)
plt.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, x2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8,3.5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8.25,3.5)
plt.subplots_adjust(wspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3,6)
plt.subplots_adjust(hspace=0.3)
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, x2)
ax1.set_title('График x**2')
ax1.set_xlim(-7, 7)
ax2.plot(x, x3)
ax2.set_title('График x**3')
ax3.plot(x, y)
ax3.set_title('График sin')
ax4.plot(x, z)
ax4.set_title('График cos')
fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
