```python
def reduce_mem_usage(df):
    start_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))
    
    for col in df.columns:
        col_type = df[col].dtype
        if (col_type != object) and (str(col_type) != 'category'):
            c_min = df[col].min()
            c_max = df[col].max()
            
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        elif col_type == object:
            df[col] = df[col].astype('category')
    
    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df
```

```python
class DataPipeline:
    '''Подготовка исходных данных'''
    
    def __init__(self):
        '''Параметры класса'''
        self.medians = None
        self.longitude_median = None
        self.latutide_median = None
        
    def fit(self, df):
        '''Сохранение статистик'''
        
        # Рассчет медиан
        self.medians = df[['population', 'housing_median_age', 'total_bedrooms']].median()
        self.longitude_median = df['longitude'].median()
        self.latitude_median = df['latitude'].median()
        
    def transform(self, df):
        """Трансформация данных"""
        
        # Пропуски
        df[['population', 'housing_median_age', 'total_bedrooms']] =\
            df[['population', 'housing_median_age', 'total_bedrooms']].fillna(self.medians)
        
        # выбросы (outliers)
        df.loc[df['longitude'] > 0, 'longitude'] = df.loc[df['longitude'] > 0, 'longitude'] * -1
        df.loc[df['longitude'] == 0, 'longitude'] = self.longitude_median
        df.loc[(df['latitude'] <= 0) | (df['latitude'] > 50), 'latitude'] = self.latutide_median
        
        # новые фичи
        df['bedroom_share'] = df['total_bedrooms'] / df['total_rooms'] * 100
        df['population_per_room'] = df['population'] / df['total_rooms']
        
        df = pd.concat([df, pd.get_dummies(df['ocean_proximity'])], axis=1)
        
        return df
```

```python
# разбиение на train / test
X = df[feature_names]
y = df[target_name]
# shuffle - случайное перемешивание строк - если в вашей бд идут сначала дорогие а потом дешевые, в итоге в трейне 
# окажутся дорогие а в тесте дешевые, поэтому нужно перемешивать
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.33, shuffle=True, random_state=42)
```

```python
# Построение моделей
# Бизнес метрика - R^2, root Mean Squared Error (RMSE)
# Бизнес метрика интерпретируема и понятна бизнесу
# ее мы показываем как результат нашей модели на бизнес встречах в презентациях
# RMSE - на сколько в среднем $ мы ошибаемся при предсказании цен
# Обычно R^2 принимает значения от 0 до 1, если 1 все предсказания идеальны Можно сказать что это точность модели
# В очень плохих моделях R^2 может быть < 0
#
# ML метрика Mean Squared Error MSE
# MSE прямопропорциональная RMSE и R^2
# MSE удобно оптимизировать (брать производные)
# MSE используется под капотом наших ML моделей

# лучшая модель - модель с наименьшей MSE на test
# MSE = 1/n * SUM((y - y_pred)^2)
# RMSE = SQRT(MSE)
# R^2 =  1 - (1/n * SUM((y - y_pred)^2)) / (1/n * SUM((y - y_mean)^2))
# наша модель учится оптимизировать MSE а мы показываем бизнесу RMSE R2 у этих 3 метрик есть хорошее свойство
# их оптимм находится в одной точке. Например min MSE = min RMSE. поэтому оптимизация mse равно оптимизации R2 and RMSE
```

```python
def evaluate_preds(true_values, pred_values, save=False):
    '''Оценка качества модели и график preds vs true'''
    print("R2:\t" + str(round(r2(true_values, pred_values), 3)) + "\n" +
         "RMSE:\t" + str(round(np.sqrt(mse(true_values, pred_values)), 3)) + "\n" + 
         "MSE:\t" + str(round(mse(true_values, pred_values), 3)))
    
    plt.figure(figsize=(8,8))
    sns.scatterplot(x=pred_values, y=true_values)
    plt.plot([0, 500000], [0, 500000], linestyle='--', color='black') # диагональ где true_values = pred_value
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('True vs Predicted values')
    
    if save == True:
        plt.savefig(REPORTS_FILE_PATH + 'report.png')
    plt.show()
```

```python
# Linear Regression
# price_pred = a0 + a1 * households + a2* rooms
# минимизируется функция потерь:
# Loss = MSE = 1/n*SUM(price - price_pred)^2
# метод наименьших квадратов
# ошибку нужно минимизировать и для этого считаем производные чтобы минимизировать loss и находим параметры для нахождения 
# минимальных коэффициентов для уменьшения ошибки.

# Назодим производные по параметрам a0, a1, a2 и решаем систему уравнений  с 3 мя неивестными a0 a1 a2
# d(Loss)/d(a0) = 1/n * SUM (2 * (price - (a0 + a1 * households + a2* rooms)) * 1) = 0
# d(Loss)/d(a1) = 1/n * SUM (2 * (price - (a0 + a1 * households + a2* rooms)) * households) = 0
# d(Loss)/d(a2) = 1/n * SUM (2 * (price - (a0 + a1 * households + a2* rooms)) * rooms) = 0

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

y_train_preds = lr_model.predict(X_train)
# постпроцессинг - мы значем что цены квартир от 100000 до 500000, и если прогноз выходит за пределы нашего результата
# то надо интервалы вернуть. Прогноз может выходить за диапазон.
y_train_preds = np.clip(y_train_preds, a_min=10000, a_max=500000)

evaluate_preds(y_train, y_train_preds)
```

```python
# мы видим что для больших цен мы делаем недопрогноз так как значения при высоких ценах смещаются от диагонали. что это может
# значить? например в банке приходят клиенты которые говорят что я хочу купить квартиру которая стоит 400000 дайте мне кредит 
# на 400000. Наша модель ответит нет та квартира стоит не 400000 а 300000 - могу дать только 300000. 
# Клиент уйдет в другой банк где ему дадут этот кредит. Это очень плохо, мы недопрогнозируем. Модель линейной регрессии
# не очень хорошая. Смотрим  метрики - R^2 = 0.612 - 61 процент. в среднем на прогнозируем мы ошибаемся на 60000 
```

```python
# измерим качество на тесте
y_test_preds = lr_model.predict(X_test)

# постпроцессинг
y_test_preds = np.clip(y_test_preds, a_min=10000, a_max=500000)
evaluate_preds(y_test, y_test_preds)
```

```python
# оценка метрики совпадает с train это хорошо, значит модель не переобучилась. Переобучение это когда ваша модель хорошо 
# предсказывает на train потому что она запомнила все значения на train но на тесте предсказывает гораздо хуже, т.е. модель
# не умеет обобщаться. в нашем случае недообучение - когда модель смогла выучить какие то зависимости и регулярно 
# недопрогнозирует каке то квартиры.
# на практике све считает сложнее
```

```python
# на примере задачи - сколько времени человек трати на онлайн игры 0 - 24 часа в день
# в ML используются бинарные деревья
# изначально у нас есть датасет с фичами. алгоритм пробегает по всем фичам и его значениям и выбирает такое разделение
# Разделение выбирается всегда меньше чем какое либо значение
# Дерево строиться до листовой величины. в листе мы берем среднее если там общая группа значений
# Дерево это набор правил на фичах
```

```python
dt_model = DecisionTreeRegressor(criterion='mse',
                                max_depth=20, # глубина дерева
                                min_samples_leaf=30, # минимальное кол-во наблюдейний в листе дерева
                                random_state=42)
# Гиперпараметры это параметры которые мы указываем вручную
# Глубина дерева от наблюдений зависит логарифмически
# пороги выбираются так чтобы минимизировать метрику
# по какому признаку проводить разделение и какой порог выбрать для этого разделения. Что мы делаем - мы генерируем все 
# возможные признаки со всеми возможными разбиениями. После этого делается прогноз вправо и влево. мы смотрим какой 
# прогноз дает наименьший MSE, и смотрим какое разбиение давало нам такой прогноз. Разбиение которое давало прогноз с 
# наименьшим MSE берем его. В продакшн библиотеках све считается чуть сложнее но насчет на быстродействие

# Глубину дерева нужно ограничивать. Если ее не ограничивать то оно будет строиться до самого конца, т.е. до тех пор 
# пока в каждом листе не останется по одному наблюдению . это не очень удачно. Потому что если в каждом листе одно 
# наблюдение мы считаем среднее по одному наблюдению. Нужно выбирать какое то кол во точек чтобы посчитать среднее.
# В итоге появится проблема переобучения - т.е. мы выучили для каждого человека значение а какое было, на на тесте с
# новыми людьми мы будем плохо прогнозировать.
dt_model.fit(X_train, y_train)

y_train_preds = dt_model.predict(X_train)
evaluate_preds(y_train, y_train_preds)
```

```python
y_test_preds = dt_model.predict(X_test)
evaluate_preds(y_test, y_test_preds)
```

```python
# Настройка и оценка финальной модели
# кросс валидация проходит на трейне test не троем
# кросс валидация и подбор гиперпараметров
# чтобы корректо подбирать гиперпараметры нам нужно кросс валидация
# у нас были данные и мы случайным образом поделили на train и test. чтобы подбирать гиперпараметры, нужно делать очень
# устойчивю кросс валидацию, те.е очень устойчиво измерять качество прогноза. на тесте мы измерям качество финальной модели
# на трейне нам надо как то подобрать гиперпараметры, подобрать фичи. чтобы на трейне это делать есть такая процедура
# кросс валидация - самая известкая KFold cross validation. Валидация по K фолдам.  
# не всегда нужно делить на случаный сеты, иногда нужно делить на временные.
# 1 шаг - выбрать количесвто фолдов. как правило их ставят равным 5 это самый обычный параметр
# 2 шаг у нас имеется K итераций - 5 у нас есть train и мы разбиваем train на случайную подвыборку на 20 процентов чтобы 
# их было 5. В первой итерации мы говорим что наша первая подвыборка это тест. давайте обучим модель на этом train
# спрогнозируем на маленьший тест и посчитаем качество прогноза - R^2.
# далее пойдет 2 итерация и мы скажем что наща 2 подвыборка это тест, теперь мы обучится на всем остальном датасете кроме
# testa. сделаем прогноз на тест и посчитаем 2 R2/ В итоге у нас получится 5 R^2 мы их усредним и получим некий
# кросс валидационный R^2
```

```python
params = {'n_estimators': [50, 100, 200, 400],
         'max_depth': [3, 5, 7, 10]}

gs = GridSearchCV(lgb_model, params, 
                 scoring='r2', # метрика
                 cv=KFold(n_splits=5, # k кол-во разбиений/итераций в кросс валидации
                         random_state=21,
                         shuffle=True),
                 n_jobs=-1 # кол-во потоков
                 )
gs.fit(X_train, y_train) # на train
```
